```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Ackley Function
def ackley(X, Y, a=20, b=0.2, c=2*np.pi):
    return -a * np.exp(-b * np.sqrt(0.5 * (X**2 + Y**2))) - np.exp(0.5 * (np.cos(c*X) + np.cos(c*Y))) + a + np.exp(1)

# 2. Styblinski-Tang Function
def styblinski_tang(X, Y):
    return 0.5 * ((X**4 - 16*X**2 + 5*X) + (Y**4 - 16*Y**2 + 5*Y))

# 3. Rastrigin Function
def rastrigin(X, Y, A=10):
    return A*2 + (X**2 - A * np.cos(2 * np.pi * X)) + (Y**2 - A * np.cos(2 * np.pi * Y))

# Meshgrid setup
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)

# Function evaluations
Z_ackley = ackley(X, Y)
Z_st = styblinski_tang(X, Y)
Z_rastrigin = rastrigin(X, Y)

# Plotting function
def plot_3d(X, Y, Z, title):
    fig = plt.figure(figsize=(10,7))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.9)
    ax.contour(X, Y, Z, zdir='z', offset=np.min(Z)-2, cmap='viridis', linestyles="solid")
    ax.set_title(title, fontsize=14)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Cost')
    fig.colorbar(surf, shrink=0.5, aspect=10)
    plt.tight_layout()
    plt.show()

# Plot Ackley
plot_3d(X, Y, Z_ackley, '3D Ackley Function')

# Plot Styblinski–Tang
plot_3d(X, Y, Z_st, '3D Styblinski–Tang Function')

# Plot Rastrigin
plot_3d(X, Y, Z_rastrigin, '3D Rastrigin Function')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Custom Function
def custom_function(X, Y):
    return (X**2 + Y**2) * np.sin(X) + np.log(1 + X**2 + Y**2)

# Meshgrid
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)

# Function Evaluation
Z_custom = custom_function(X, Y)

# Plotting
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z_custom, cmap='coolwarm', edgecolor='none', alpha=0.9)
ax.contour(X, Y, Z_custom, zdir='z', offset=np.min(Z_custom)-2, cmap='coolwarm', linestyles="solid")
ax.set_title(r"$f(x, y) = (x^2 + y^2) \sin(x) + \log(1 + x^2 + y^2)$", fontsize=14)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Cost')
fig.colorbar(surf, shrink=0.5, aspect=10)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Schwefel function
def schwefel(x, y):
    return 418.9829 * 2 - (x * np.sin(np.sqrt(np.abs(x))) + y * np.sin(np.sqrt(np.abs(y))))

# Create a meshgrid
x_vals = np.linspace(-500, 500, 400)
y_vals = np.linspace(-500, 500, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = schwefel(X, Y)

# Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', linewidth=0, antialiased=True)

ax.set_title("Schwefel Function Surface Plot", fontsize=14)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("f(X, Y)")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Eggholder function definition
def eggholder(x, y):
    return -(y + 47) * np.sin(np.sqrt(abs(x/2 + (y + 47)))) \
           - x * np.sin(np.sqrt(abs(x - (y + 47))))

# Create a grid of points
X = np.linspace(-512, 512, 500)
Y = np.linspace(-512, 512, 500)
X, Y = np.meshgrid(X, Y)
Z = eggholder(X, Y)

# Plotting
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis', linewidth=0, antialiased=True)

ax.set_title('Eggholder Function', fontsize=16)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X,Y)')

fig.colorbar(surf, shrink=0.5, aspect=10)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Michalewicz function definition
def michalewicz(X, m=10):
    x, y = X
    term1 = np.sin(x) * (np.sin((x**2)/np.pi) ** (2 * m))
    term2 = np.sin(y) * (np.sin((2 * y**2)/np.pi) ** (2 * m))
    return -(term1 + term2)

# Grid setup
x = np.linspace(0, np.pi, 500)
y = np.linspace(0, np.pi, 500)
X, Y = np.meshgrid(x, y)
Z = michalewicz((X, Y), m=10)

# Plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

ax.set_title('Michalewicz Function (m=10)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Rosenbrock function
def rosenbrock(x, y, a=1, b=100):
    return (a - x)**2 + b * (y - x**2)**2

# Create grid
x = np.linspace(-2, 2, 400)
y = np.linspace(-1, 3, 400)
X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)

# Plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Labels
ax.set_title('Rosenbrock Function', fontsize=16)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define Griewank function
def griewank(x, y):
    term1 = (x**2 + y**2) / 4000
    term2 = np.cos(x / np.sqrt(1)) * np.cos(y / np.sqrt(2))
    return term1 - term2 + 1

# Meshgrid for plotting
X = np.linspace(-10, 10, 400)
Y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(X, Y)
Z = griewank(X, Y)

# Plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', linewidth=0, antialiased=True)

ax.set_title('Griewank Function', fontsize=14)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X,Y)')

fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Levy N.13 function
def levy(x, y):
    w1 = 1 + (x - 1) / 4
    w2 = 1 + (y - 1) / 4
    term1 = np.sin(np.pi * w1) ** 2
    term2 = ((w1 - 1) ** 2) * (1 + 10 * np.sin(np.pi * w1 + 1) ** 2)
    term3 = ((w2 - 1) ** 2) * (1 + np.sin(2 * np.pi * w2) ** 2)
    return term1 + term2 + term3

# Meshgrid
X = np.linspace(-10, 10, 400)
Y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(X, Y)
Z = levy(X, Y)

# Plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='plasma', linewidth=0, antialiased=True)

ax.set_title("Lévy N.13 Function", fontsize=14)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("f(X,Y)")
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Himmelblau function definition
def himmelblau(x, y):
    return (x**2 + y - 11)**2 + (x + y**2 - 7)**2

# Create grid
x = np.linspace(-6, 6, 400)
y = np.linspace(-6, 6, 400)
X, Y = np.meshgrid(x, y)
Z = himmelblau(X, Y)

# 3D surface plot
fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.9, edgecolor='none')
ax1.set_title("Himmelblau Function - 3D Surface")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("f(x, y)")

# Mark known minima
minima = [(3.0, 2.0), (-2.805118, 3.131312),
          (-3.779310, -3.283186), (3.584428, -1.848126)]
for m in minima:
    ax1.scatter(m[0], m[1], himmelblau(m[0], m[1]), color='red', s=50)

# 2D contour plot
ax2 = fig.add_subplot(1, 2, 2)
contour = ax2.contourf(X, Y, Z, levels=50, cmap='viridis')
ax2.set_title("Himmelblau Function - Contour")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
plt.colorbar(contour, ax=ax2)

# Mark minima on contour
for m in minima:
    ax2.plot(m[0], m[1], 'ro')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Langermann function definition
def langermann(x, y):
    a = np.array([3, 5, 2, 1, 7])
    b = np.array([5, 2, 1, 4, 9])
    c = np.array([1, 2, 5, 2, 3])
    m = len(a)

    total = 0
    for i in range(m):
        r2 = (x - a[i])**2 + (y - b[i])**2
        total += c[i] * np.exp(-r2 / np.pi) * np.cos(np.pi * r2)
    return total

# Create grid
x = np.linspace(0, 10, 400)
y = np.linspace(0, 10, 400)
X, Y = np.meshgrid(x, y)
Z = langermann(X, Y)

# Create figure
fig = plt.figure(figsize=(12, 5))

# 3D surface plot
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.9)
ax1.set_title("Langermann Function - 3D Surface")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("f(x, y)")

# 2D contour plot
ax2 = fig.add_subplot(1, 2, 2)
contour = ax2.contourf(X, Y, Z, levels=50, cmap='viridis')
ax2.set_title("Langermann Function - Contour")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
plt.colorbar(contour, ax=ax2)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Easom function
def easom(x, y):
    return -np.cos(x) * np.cos(y) * np.exp(-((x - np.pi)**2 + (y - np.pi)**2))

# Grid
x = np.linspace(-100, 100, 400)
y = np.linspace(-100, 100, 400)
X, Y = np.meshgrid(x, y)
Z = easom(X, Y)

# 3D surface plot
fig = plt.figure(figsize=(12, 5))

ax = fig.add_subplot(1, 2, 1, projection="3d")
ax.plot_surface(X, Y, Z, cmap="viridis", edgecolor="none", alpha=0.9)
ax.set_title("Easom Function (3D)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x,y)")

# Contour plot
ax2 = fig.add_subplot(1, 2, 2)
contour = ax2.contourf(X, Y, Z, levels=50, cmap="viridis")
fig.colorbar(contour, ax=ax2)
ax2.set_title("Easom Function (Contour)")
ax2.set_xlabel("x")
ax2.set_ylabel("y")

plt.tight_layout()
plt.show()
```
